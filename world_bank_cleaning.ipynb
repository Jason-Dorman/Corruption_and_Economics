```python
# load poverty headcount dataset
poverty_headcount = pd.read_csv('poverty_headcount.csv')

poverty_headcount_df = pd.DataFrame(poverty_headcount)

poverty_headcount_df.head()
```

```python
# load debt dataset
debt = pd.read_csv('debt.csv')

debt_df = pd.DataFrame(debt)

debt_df.head()
```

```python
# clean columns
#poverty_headcount_df = poverty_headcount_df.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]' , '2000 [YR2000]'])

revenue_df = revenue_df.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]' , '2000 [YR2000]'])

debt_df = debt_df.drop(columns=['Series Name', 'Series Code', '1990 [YR1990]' , '2000 [YR2000]'])
```

```python
# check countries for poverty headcount
p_unique_countries = poverty_headcount_df['Country Name'].unique()
p_unique_countries
```

```python
# check number of countries for poverty headcount
len(p_unique_countries)
```

```python
# keep only the countries for poverty, revenue, debt
debt_df = debt_df.iloc[:217]

poverty_headcount_df = poverty_headcount_df.iloc[:217]

revenue_df = revenue_df.iloc[:217]
```

```python
# replace cells with no value with NaN
debt_df = debt_df.replace('..', 'NaN')

poverty_headcount_df = poverty_headcount_df.replace('..', 'NaN')

revenue_df = revenue_df.replace('..', 'NaN')
```

```python
# rename the columns
poverty_headcount_df = poverty_headcount_df.rename(columns={"Country Name":"country_name", "Country Code":"country_code", "2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017" , "2018 [YR2018]":"2018"})

revenue_df = revenue_df.rename(columns={"Country Name":"country_name", "Country Code":"country_code", "2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017" , "2018 [YR2018]":"2018"})

debt_df = debt_df.rename(columns={"Country Name":"country_name", "Country Code":"country_code", "2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017" , "2018 [YR2018]":"2018"})
```

```python
# clean datasets
clean_poverty_headcount_df = poverty_headcount_df.melt(id_vars=['country_name', 'country_code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                    , var_name ='Year', value_name ='Poverty Headcount Ratio')

clean_revenue_df = revenue_df.melt(id_vars=['country_name', 'country_code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                    , var_name ='Year', value_name ='revenue')

clean_debt_df = debt_df.melt(id_vars=['country_name', 'country_code'], value_vars=['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']\
                    , var_name ='Year', value_name ='central_government_debt')
```

```python
# check poverty dataframe
clean_poverty_headcount_df.head()
```

```python
# check revenue dataframe
clean_revenue_df.head()
```

```python
# clean corruption colums
#corruption_df = corruption_df.drop(columns=['Indicator Code'])
corruption_df = corruption_df.drop(columns=['1996', '1998', '2000'])
corruption_df.head()
```

```python
# clean corruption data
clean_corruption_df = corruption_df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name'], value_vars=['2002', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']\
                    , var_name ='Year', value_name ='Indicator Value')

clean_corruption_df.head()
```

```python
# rename columns for corruption dataset
clean_corruption_df = clean_corruption_df.rename(columns={"Country Name":"country_name", "Country Code":"country_code", "Indicator Name":"indicator_name", "Indicator Value":"indicator_value"})

clean_corruption_df.head()
```

```python
#check datatypes for revenue
clean_revenue_df.dtypes
```

```python
# check datatypes for debt
clean_debt_df.dtypes
```

```python
# check datatypes for poverty
clean_poverty_headcount_df.dtypes
```

```python
# change datatypes
clean_revenue_df = clean_revenue_df.astype({"revenue": float})

clean_debt_df = clean_debt_df.astype({"central_government_debt": float})

clean_poverty_headcount_df = clean_poverty_headcount_df.astype({"Poverty Headcount Ratio": float})
```

```python
# merge debt, revenue, poverty
merged_indicator_df = clean_debt_df.merge(clean_poverty_headcount_df, on=['country_name'], how='inner')\
                       .merge(clean_revenue_df, on=['country_name'], how='inner')

merged_indicator_df.head()
```

```python
# clean columns for merged df
merged_indicator_df = merged_indicator_df.drop(columns=['country_code_x', 'Year_x', 'country_code_y' , 'Year_y'])
merged_indicator_df.head()
```
